```python
# 1. numpy : 난수발생
print(dir(np.random))
print()

data = np.random.rand() # 0<= random < 1사이의 실수 한개
print(type(data), data)

# np.random.rand(d0,d1....)
data = np.random.rand(50) # 1차원으로 50개의 난수
print(type(data), data)

data = np.random.rand(2,3) # 2차원 즉 2행 3열의 난수발생
print(type(data), data)
```

```python
# 2. pandas

# 1) 1차원자료
# pd.Series?
data = np.random.rand(50)
s = pd.Series(data)
print(type(s), s)
```

```python
s.plot(kind='line')
# s.plot?
```

```python
# 2) 2차원자료생성
# pd.DataFrame?
# a. numpy로 10행 3열의 2차원배열(정확하게는 행렬 or matrix)
data = np.random.rand(10,3)
print(type(data),data)
```

```python
# b. 행렬을 DataFrame으로 변환
df = pd.DataFrame(data)
print(type(df), df)
df
print(df)
display(df)
```

```python
# c. df의 컬럼명을 정하기
df = pd.DataFrame(data, columns=['2019년', '2020', 2021])
df
```

```python
# d. df 시각화
df.plot()
df.plot?
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='barh')
```

```python
df.plot(kind='area')  # 기본:누적, stacked=True
df.plot(kind='area', stacked=False) # 비누적
```

```python
# pie chart
# 5개의 난수를 발생, Series자료형으로 Pie chart로 시각화
np.random.seed(42)
s = pd.Series(data=np.random.rand(5), index=['2019','2020','2021','2022','2023'])
print(s)

# 시각화
s.plot(kind='pie')
```

```python
# autopct : 퍼센트표시(기본 소숫점2자리)
s.plot(kind='pie', autopct='%.2f%%', figsize=(2,2), fontsize=5)
```
