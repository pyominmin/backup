# A. 리스트(list)자료형

* 파이썬에서 리스트 자료형을 만들 때는 `대괄호[]`로 선언하고 그 안에 요소의 값을 콤마(,)로 구분한다.
* 파이썬에서는 리스트자료형은 String(문자열, str)만큼 자주 사용하는 자료형이다.

>리스트변수명 = [값1,...값n]

## 리스트란?

1. 순서가 있는 수정이 가능한 `객체들의 집합`이다.
2. 수정, 삭제, 추가가 가능하다. 즉, mutable자료형이다.
3. list는 `대괄호[]로 선언하고 내부의 요소(원소, 값...)는 컴마(,)로 구분`한다.
4. list의 순서(index)는 `0부터 시작`한다.
5. index의 범위가 벗어나면 에러가 발생된다.
6. list에는 파이썬의 어떠한 자료형도 저장할 수 있다.
7. 즉, 숫자,문자열,함수,튜플,리스트등의 모든 자료형의 값을 요소로 저장할 수 있다.
8. list는 `마지막요소뒤에 컴마가 있어도 에러가 발생되지 않는다`
9. 보통은 마지막의 컴마는 생략하지만 편의를 위해서 컴마를 찍기도 한다.

### 1. list의 인덱싱/슬라이싱

```python
a = [1, 'a', [5,6,7,[7,8,9,[10,11,12]]], print, (), {}, {1}] 

# 1. 인덱싱
print(a[2][3][3][1])
print(a[0], a[1], a[2], a[3])
print(type(a[0]), type(a[1]), type(a[2]), type(a[3]), type(a[4]), type(a[5]), type(a[6]))
print()

# 1) 인덱스는 음수도 사용, 음수인덱스는 뒤에서부터 순서처리
print(type(a[-1]))
print(type(a[-2]))

# 2) list의 요소 사용하기
# print(a[0] + a[1]) # TypeError: unsupported operand type(s) for +: 'int' and 'str'

# list는 javascript의 array와 비슷한 자료형이다.
a[-4]
```

```python
# 2. list의 슬라이싱
# ... 문자열과 동일하게 list자료형도 slicing기법을 사용할 수 있다.
# ... 사용법 : list[start [,end-1, [step=1]]]의 형태로 사용한다.

# 1) 1차원 슬라이싱
a = [1,2,3,4,5,]
print(a[:], a[0:], a[:5])
print(type(a[0:1]), a[0:1])
print(type(a[0:2]), a[0:2])
print(type('12345'[0:2]), '12345'[0:2])
```

```python
# 2) 다차원슬라이싱 - 중첩된 list자료형의 슬라이싱
a = [1,2,3,['a','b','c'],4,5]

# a에서 c를 추출하기
print(a)
print(a[2:5])
print(a[3:4])  # 슬라이싱 - 결과가 2차원 list
print(a[3])    # 인덱싱   - 결과가 1차원 list
print()

# print(a[3:4][2]) # IndexError
print('c값 추출하기 = ', a[3][2])
```

### 3. list 수정 및 삭제

```python
# 1. list에서 한 개의 값을 수정하기
a = [1,2,3]
a[1] = 'x'
print(a)
```

```python
# 2. list에서 연속된 범위의 값을 수정하기
a = [1,2,3]
print(a[1:2])
a[1:2] = ['x', 'y', 'z']
print(a)

print(a[1:4])
a[1:4] = ['@', '#', '$', '!', '%', '*', '&']
print(a)
```

```python
# 4. list의 step옵션사용하기
# list[:] = 슬라이싱, list[start:end-1:step]
sample = 'show how to index into sequences'.split()
print(type(sample), sample)
print(sample[:])
print(sample[::])
print(sample[::1])
print(sample[::2])
print(sample[::3])
print(sample[::4])
print(sample[::5])
print(sample[::6])
print()

# step옵션 음수 정의가능
print(sample[::-1])
print(sample[::-2])
print(sample[::-3])
print(sample[::-4])
print(sample[::-5])
```

```python
# 1. append : list에 요소 추가하기
a = [1,2,3]
print(a)

a.append(4)
print(a)

a.append([4,5])  # 한 개의 요소값으로 추가
print(a)

a.extend([6,7]) # 각각의 요소값으로 추가
print(a)

a.append((8,9)) # 한 개의 요소값(tuple형)으로 추가
print(a)

a.extend((10,11)) 
print(a)

a.append({'name':'홍길동', 'age':1000}) 
print(a)
```

```python
# 2. sort()
# ... 정렬은 요소의 데이터타입이 동일해야 정렬할 수 있다.(in-place)
# ... 데이터타입이 다를 경우 에러 발생
# ?a.sort
a = [1,4,3,2]
result = a.sort() # sort는 return이 None
print(a, result)

b = ['x','z','y']
print(b.sort())
print(b)

# 데이터타입이 다를 경우는 에러발생
c = ['x','z','y', 1,4,3,2]
# c.sort() # TypeError: '<' not supported between instances of 'int' and 'str'
```

```python
# 3. reverse() : 현재상태에서 역순으로 뒤집기(정렬하는 것이 아니라 뒤집기만 수행), in-place
# c.reverse?
print(c)
c.reverse()
```

```python
# 4. index(문자열)
# a.index?
a = [1,2,'a',3,'a','b']
print(a.index('a'))
# print(a.index('x')) # 찾는 문자가 없을 경우 에러

# a.index(value, start=0, stop=9223372036854775807, /)
# start는 시작위치를 정의
# stop은 종료위치를 정의
print(a.index('a', 3))
print(a.index('a', 3, 1000))
```

```python
# 5. remove(전달값): 매개변수로 전달된 요소를 삭제
# 만약, 여러개의 값이 존재할 경우 맨 앞의 값만 삭제
# 삭제할 값이 없을 경우 에러발생
# 사용법 : list명.remove(삭제할 값)
# a.remove?
a = [1,2,3,1,2,3]
a.remove(3)
print(a)

# a.remove(10) ValueError: list.remove(x): x not in list
```

```python
# 6. pop(매개변수) : 매개변수의 위치에 해당하는 요소를 읽어온 후 삭제
# 매개변수를 지정하지 않았다면 맨 마지막에 위치한 요소를 읽은 후 삭제
# 인덱스범위를 벗어나는 경우에는 에러발생
# 사용법 : list명.pop()
# a.pop?
a = [1,2,3,4]
print(a.pop(), a)
print(a.pop(1), a)

# a.pop(10) IndexError: pop index out of range
```

```python
# 7. count(값) : list요소값의 갯수를 리턴
# 사용법 : list명.count(값)
# a.count?
a = [1,2,3,1,2,3,2,4,2]
print('1의 갯수 = ', a.count(1))
print('2의 갯수 = ', a.count(2))
print('5의 갯수 = ', a.count(5))
```

```python
# 8. extend(iterable) : list를 확장, iterable자료형(문자열, list, tuple..)만 전달 가능
# a.extend?
a = [1,2,3]
a.append([4,5])
print(a)

a = [1,2,3]
a.extend([4,5])
print(a)

a.extend('hello')
print(a)

a.extend({1,2}) # set 자료형
print(a)

a.extend({'name':'홍길동', 'age':1000}) # dict 자료형, key만 확장
print(a)
```

```python
# 9. insert(index, object) : 전달한 값을 list에 삽입(추가)
# a.insert?
a.insert(2, 'hong')
print(a)

a.insert(10, '홍길동')
print(a)
```

```python
# 10. clear() : 요소전체를 삭제
# a.clear?
print(len(a), a)
a.clear()
print(len(a), a)
```

```python
# 11. copy() : 리스트를 복사, 리스트는 객체(java의 참조타입객체)
# 파이썬에서 모든 자료형은 객체이다. 즉, 객체의 주소를 참조하고 있다.
a = 10;
# print(dir(a))
# id(객체) - 객체가 저장되어 있는 메모리주소(java의 hash값)
print('a가 저장되어 있는 메모리주소 = ', id(a))

a = [1,2,3,['x','y']]
print('a가 저장되어 있는 메모리주소 = ', id(a))
print('a[0]가 저장되어 있는 메모리주소 = ', id(a[0]))
print('a[3]가 저장되어 있는 메모리주소 = ', id(a[3]))
print()

# 복사 : 얕은 복사와 깊은 복사
# 1) 얕은 복사
b = a
print(id(a), id(b))
print(id(a[0]), id(b[0]))
print(id(a[3]), id(b[3]))
print()

a[0] = 'hong'
print(a[0], b[0])
print(id(a[0]), id(b[0]))

a[3] = ['a','b']
print(a, b)
print(a[3], b[3])
print(id(a[3]), id(b[3]))
print()
```

```python
# 2) 깊은 복사
c = a[:]
print(a, c)
print(id(a), id(c))

d = a.copy()
print(id(a), id(d))
```

```python
import copy as cp
print(dir(cp))

e = cp.copy(a)
print(id(a), id(e))
print(id(a[3]), id(e[3]))
print()

a[3] = ['@', '#']
print(a, e)
print(id(a[3]), id(e[3]))
```

```python
f = cp.deepcopy(a)
print(a, f)
print(id(a), id(f))
print(id(a[3]), id(f[3]))
```

```python
# 객체의 비교
print(id(a), id(b))
print(id(a), id(c))
print(id(a), id(d))
print(id(a), id(e))
print(id(a), id(f))
print()

# == : 객체의 주소만 비교
# is : 객체의 주소와 값을 비교, java에서는 equals(), javascript에서는 ===
print('a==b', a==b, a, b)
print('a==c', a==c, a, c)
print('a==d', a==d, a, d)
print('a==e', a==e, a, e)
print('a==f', a==f, a, f)
print()

print('a is b', a is b, a, b)
print('a is c', a is c, a, c)
print('a is d', a is d, a, d)
print('a is e', a is e, a, e)
print('a is f', a is f, a, f)
```
